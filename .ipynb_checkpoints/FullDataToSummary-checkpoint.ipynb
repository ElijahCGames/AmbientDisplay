```python
# Take each csv
# Group by stn, month, year
# Average 3 months into one season

import pandas as pd
import numpy as np
```

```python
stations = pd.read_csv("stations.csv")
stations = stations.set_index("name")

seasons = { 1:"Winter",
            2:"Winter",
            3:"Spring",
            4:"Spring",
            5:"Spring",
            6:"Summer",
            7:"Summer",
            8:"Summer",
            9:"Fall",
            10:"Fall",
            11:"Fall",
            12:"Winter"}

def makeSeason(year):
    print(year)
    se = pd.read_csv("weather" + str(year) + ".csv")
    se = se.join(stations,on="name")
    se["year"] = se["year"] + se["mo"] // 12
    se  = se[se["year"] <  2020]
    se["season"] = se["mo"].map(seasons)
    return se.groupby(["CommonName","year","season"]).mean()[["temp","max","min","prcp"]]

def makeSeasonSum(year):
    print(year)
    se = pd.read_csv("weather" + str(year) + ".csv")
    se = se.join(stations,on="name")
    se["year"] = se["year"] + se["mo"] // 12
    se  = se[se["year"] <  2020]
    se["season"] = se["mo"].map(seasons)
    return se.groupby(["CommonName","year","season"]).sum()[["temp","max","min","prcp"]]
```

```python
df = pd.DataFrame()
df2 = pd.DataFrame()
for n in range(1960,2020,10):
    df = df.append(makeSeason(n))
    df2 = df2.append(makeSeasonSum(n))
df = df[~df.index.duplicated(keep='last')]
df2 = df2[~df2.index.duplicated(keep='last')]
#df.to_csv("seasonalWeather.csv")
```
